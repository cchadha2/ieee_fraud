```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
pd.set_option('display.max_columns', 500)
SEED = 5000
```

```python
train_ids = pd.read_csv('data/train_identity.csv.zip')
train_trans = pd.read_csv('data/train_transaction.csv.zip', nrows=100000)
```

```python
train = pd.merge(train_trans, train_ids, on='TransactionID', how='inner')
```

```python
train.head(10)
```

```python
print(train_trans.shape)

print(train_ids.shape)

print(train.shape)
```

```python
categorical_features=['ProductCD', 'card1', 'card2', 'card3', 'card4', 'card5', 'card6',
                      'addr1', 'addr2', 'P_emaildomain', 'R_emaildomain', 'M1',
                      'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9',
                      'DeviceType', 'DeviceInfo', 'id_12', 'id_13', 'id_14', 'id_15',
                      'id_16', 'id_17', 'id_18', 'id_19', 'id_20', 'id_21', 'id_22',
                      'id_23', 'id_24', 'id_25', 'id_26', 'id_27', 'id_28', 'id_29',
                      'id_30', 'id_31', 'id_32', 'id_33', 'id_34', 'id_35', 'id_36',
                      'id_37', 'id_38']
for feature in categorical_features:
    train[feature] = train[feature].astype('category')
```

```python
train_encoded = pd.get_dummies(train, dummy_na=True)
```

```python
train_encoded.shape
```

```python
train_encoded.head()
```

```python
X_train = train_encoded.drop('isFraud', axis=1)
y_train = train_encoded['isFraud']
```

```python
# X_train.fillna(method='backfill', inplace=True)
# X_train.fillna(method='pad', inplace=True)
```

```python
X_train.head()
```

```python
X_train = X_train.loc[:,X_train.isna().any() == False]
```

```python
X_train.head()
```

```python
log_reg = LogisticRegression()
log_reg_trained = log_reg.fit(X_train, y_train)
```

```python
coefficients = np.array([[log_reg_trained.coef_], [X_train.columns.values]]).tolist()
```
